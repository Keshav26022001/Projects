## Data Preprocessing and Feature Engineering.

```python
data.isnull().sum()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


X=data.iloc[:,:-1]
y=data.price_range
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# Log Transformation to handle very small values
#X_train = np.log1p(X_train)
#X_test = np.log1p(X_test)

```

```python
# Feature Scaling
#scaler = StandardScaler()
#X_train_scaled = scaler.fit_transform(X_train)
#X_test_scaled = scaler.transform(X_test)

```

```python
# Train the Multinomial Logistic Regression Model with increased regularization
model = LogisticRegression(multi_class='multinomial', solver='saga', max_iter=10000, C=0.1)  # Adjust C value as needed
model.fit(X_train, y_train)

```

```python
# Make Predictions
y_pred = model.predict(X_test)

```

```python

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Generate Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Plot Confusion Matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.show()

# Classification Report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```

```python
#Make predictions for training and test data
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
# Training and Test Accuracy
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print(f"Training Accuracy: {train_accuracy:.2f}")
print(f"Test Accuracy: {test_accuracy:.2f}")

```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],
    'solver': ['newton-cg', 'lbfgs', 'liblinear']
}
grid = GridSearchCV(LogisticRegression(multi_class='multinomial'), param_grid, refit=True, verbose=2)
grid.fit(X_train, y_train)
```

```python
# Make predictions with the tuned model
y_pred_test = grid.predict(X_test)

# Evaluate the tuned model
test_accuracy = accuracy_score(y_test, y_pred_test)
print(f"Test Accuracy: {test_accuracy:.2f}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.show()

# Classification Report
report = classification_report(y_test, y_pred_test)
print("Classification Report:\n", report)
```

```python
#Make predictions for training and test data
y_pred_train = grid.predict(X_train)
y_pred_test = grid.predict(X_test)
# Training and Test Accuracy
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print(f"Training Accuracy: {train_accuracy:.2f}")
print(f"Test Accuracy: {test_accuracy:.2f}")

```

```python
best_model = grid.best_estimator_
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(best_model, X_train, y_train, cv=5)
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean CV Score: {np.mean(cv_scores):.2f}")
```
