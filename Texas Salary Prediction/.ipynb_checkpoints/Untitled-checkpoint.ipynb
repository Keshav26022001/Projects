```python
# Importing Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# There is a warning while loading the dataset this is due to the dataset cotains mixed data types to overcome this
# Loading the dataset with low_memory set to False
df = pd.read_csv("salary.csv", low_memory=False)
df.head(2)
```

```python
# IQR method
Q1 = df['ANNUAL'].quantile(0.25)
Q3 = df['ANNUAL'].quantile(0.75)
IQR = Q3-Q1
lb = Q1-1.5*IQR
print(lb)
ub = Q3+1.5*IQR
print(ub)
outliers = df[(df['ANNUAL']< lb) | (df['ANNUAL']> ub)]
outliers
filtered_df = df[(df['ANNUAL'] >= lb) & (df['ANNUAL'] <= ub)]
filtered_df
```

```python
filtered_df[['HRLY RATE', 'ANNUAL']].sort_values(by='HRLY RATE')
```

```python
sns.kdeplot(filtered_df['HRLY RATE'])
```
